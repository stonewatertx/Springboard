The daily average usage clearly shows a regular pattern that seems to be in weekly manner. Also, electricity consumption seems to vary on season too. Let's convert the data to monthly average and see if it becomes more obvious.

```python
df.resample('M').mean().plot(figsize=(20,3),marker='*')
```

Electricity does peak in summer season and bottom in winter season. This is due to increased air conditioning usage as the weather gets warmer.

## 3. Decomposition on daily average energy usage

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
daily = df.resample('D').mean().meter_reading
```

```python
decomposition = seasonal_decompose(daily)
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid
```

```python
fig, ax = plt.subplots(3,1,figsize=(15,10))
plt.subplot(311)
plt.plot(daily, label = 'actual')
plt.plot(trend, label = 'trend')
# plt.plot(daily - seasonal, label = 'actual - seasonal')
plt.legend(loc = 'best')

plt.subplot(312)
plt.plot(seasonal, label = 'seasonal')
# plt.plot(daily-trend, label = 'actual - trend')
plt.legend(loc = 'best')

plt.subplot(313)
plt.plot(residual, label = 'Residuals')
plt.legend(loc = 'best')

plt.tight_layout()
```

The decomposition process divides the data into three components: trend, seasonality and residual. 

## 4. Daily electricity usage vs air temperature

```python
# our target building is in site 13 according to metadata
w = weather[weather['site_id']==13].resample('D')['air_temperature','dew_temperature'].mean()
```

```python
# join the weather data and electricity data
daily_w = pd.DataFrame(daily).join(w).dropna()
```

```python
# define a function to convert the datetime to weekday/weekend
def ww_convert(x):
    if x <=4:
        return 'Weekday'
    else:
        return 'Weekend'
```

```python
daily_w['Day'] = list(map(ww_convert, daily_w.index.dayofweek))
```

```python
daily_w.head()
```

```python
fig, ax = plt.subplots(1,2, figsize=(12,5))
sns.scatterplot(data=daily_w,x='air_temperature',y='meter_reading',hue='Day',ax=ax[0])
sns.scatterplot(data=daily_w,x='dew_temperature',y='meter_reading',hue='Day',ax=ax[1])
plt.tight_layout()
```

Looks like when air temperature and dew temperature are above a threshold, they start to impact electricity usage. We'll incorporate those features into our model in next session.
