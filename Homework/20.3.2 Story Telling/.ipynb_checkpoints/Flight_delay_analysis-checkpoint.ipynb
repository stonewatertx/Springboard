```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import math
```

# Load data

```python
raw = pd.read_csv('flight_delays_train.csv')
```

```python
raw.head()
```

```python
print('Number of Carriers: ',raw.UniqueCarrier.nunique())
print('Number of Airports: ',raw.Origin.nunique())
```

```python
raw.info()
```

```python
data = raw.copy()
```

```python
data['Month'] = data.Month.map(lambda x: x.split('-')[-1]).astype('int32')
data['DayofMonth'] = data.DayofMonth.map(lambda x: x.split('-')[-1]).astype('uint8')
data['DayOfWeek'] = data.DayOfWeek.map(lambda x: x.split('-')[-1]).astype('uint8')
```

```python
data['DepHr']=data.DepTime.apply(lambda x: math.ceil(x/100))
```

```python
data['delayed'] = pd.to_numeric(data.dep_delayed_15min.map({'Y':1,'N':0}))
```

```python
data.info()
```

```python
sns.pairplot(data)
```

```python
data.groupby('DepHr').delayed.count().plot(kind='bar',figsize=(6,4))
plt.xlabel('Departure hour')
plt.ylabel('Number of flights')
plt.title('Number of flights by departure time')
```

```python
data[data.DepTime>2400]
```

```python
data.DepHr.replace(25,1,inplace=True)
data.DepHr.replace(26,2,inplace=True)
```

```python
flight_hr = pd.DataFrame()
flight_hr['On_Schedule'] = data[data.delayed==0].groupby('DepHr').delayed.count()
flight_hr['Delayed'] = data[data.delayed==1].groupby('DepHr').delayed.count()
flight_hr['Chance_of_delay'] = flight_hr['Delayed']/(flight_hr['Delayed']+flight_hr['On_Schedule'])
```

```python
plt.subplots(figsize=(6,4))
plt.pie(data.groupby('delayed').DepHr.count(),autopct="%.1f%%",labels=['On Schedule','Delayed'])
plt.title('Delayed flight percentage')
```

```python
plt.subplots(figsize=(6,4))
flight_hr['On_Schedule'].plot(kind='bar')
flight_hr['Delayed'].plot(kind='bar',color='#ff7f0e',bottom=flight_hr['On_Schedule'])
plt.ylabel('Number of flights')
plt.xlabel('Departure hour')
plt.title('Number of delayed and on-schedule flights by hour')
plt.legend()
```

```python
plt.subplots(figsize=(6,4))
flight_hr['Chance_of_delay'].plot(marker='*')
plt.ylabel('Percentage of delayed flight')
plt.xlabel('Departure hour')
plt.title('Percentage of delayed flights by departure hour')
```

## by airport

```python
airport = pd.DataFrame(data.groupby('Origin').delayed.count())
```

```python
airport.columns=['flights']
```

```python
airport.head()
```

```python
airport = airport.join(data.groupby('Origin').delayed.mean())
```

```python
airport.head()
```

```python
airport.shape
```

```python
airport.flights.sort_values(ascending=False).head(20).plot(kind='bar',figsize=(6,4))
plt.ylabel('Number of flights')
plt.title('Top 20 airports with most flights')
```

```python
big_airport = airport[airport['flights']>200]
```

```python
big_airport
```

```python
sns.scatterplot(big_airport.flights,big_airport.delayed)
```

```python
plt.subplots(figsize=(15,5))
sns.regplot(big_airport.flights,big_airport.delayed)
plt.xlabel('Number of flights')
plt.ylabel('Percentage of delayed flights')
plt.ylim(0,0.4)
```

## by carrier

```python
carrier = pd.DataFrame(data.groupby('UniqueCarrier').delayed.count())
```

```python
carrier.columns=['flights']
```

```python
carrier.head()
```

```python
carrier = carrier.join(data.groupby('UniqueCarrier').delayed.mean())
```

```python
carrier.head()
```

```python
carrier.flights.sort_values(ascending=False).plot(kind='bar',figsize=(6,4))
plt.ylabel('Number of flights')
plt.xlabel('Carrier')
plt.title('Number of flights by carriers')
```

```python
plt.subplots(figsize=(6,4))
carrier.delayed.sort_values(ascending=False).plot(kind='bar')
plt.ylabel('Percentage of delayed flights')
plt.xlabel('Carrier')
plt.title('Percentage of delayed flights by carriers')
```

```python
plt.subplots(figsize=(6,4))
plt.scatter(carrier.flights,carrier.delayed)
```

```python
carrier['>4000 flights'] = carrier.flights.apply(lambda x: 'Yes' if x > 4000 else 'No')
```

```python
carrier
```

```python
# plt.subplots(figsize=(10,5))
sns.lmplot(data=carrier, x='flights',y='delayed',hue='>4000 flights',height=4, aspect=1.5,legend_out=False)
plt.xlabel('Number of flights')
plt.ylabel('Percentage of delayed flights')
plt.ylim(0,0.4)
```
