```python
# line plot of time series
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
```

```python
# load dataset
series = pd.read_csv('daily-min-temperatures.csv', header=0, index_col=0)
```

```python
print(series.head())
# line plot of dataset
series.plot()
```

# Training, Validation split

```python
split_point = len(series) - 7
dataset, validation = series[0:split_point], series[split_point:]
print('Dataset %d, Validation %d' % (len(dataset), len(validation)))
```

```python
# create a differenced series
def difference(dataset, interval=1):
    diff = list()
    for i in range(interval, len(dataset)):
        value = dataset[i] - dataset[i - interval]
        diff.append(value)
    return np.array(diff)
```

```python
def inverse_difference(history, forecast, interval=1):
	return forecast + history[-interval]
```

```python
X=dataset.values
differenced = difference(X, 365)
```

```python
differenced.shape, X.shape
```

```python
differenced
```

```python
differenced.shape
```

```python
plt.subplots(figsize=(15,3))
plt.plot(differenced)
```

# fit model (differenced)

```python
model = ARIMA(differenced, order=(7,0,1))
model_fit = model.fit()
```

### one-step out-of sample forecast

```python
model_forecast = model_fit.forecast(2)
```

```python
model_forecast
```

```python
forecast = inverse_difference(X, model_forecast, 365)
forecast
```

### Predict (by default, it only predict the training data set. When defining the start/stop, it can forecast)
